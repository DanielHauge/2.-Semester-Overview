```python
%%sql CREATE TABLE hello_log(
language VARCHAR,
time timestamp
);
```

```python
%%sql CREATE FUNCTION myCoolFunction()
    RETURNS trigger as
    $$
    BEGIN
        INSERT INTO hello_log(language, time) VALUES (NEW.language, now());
        RETURN null;
    END
    $$
    LANGUAGE PLPGSQL
```

```python
%%sql CREATE TRIGGER trigname
    AFTER INSERT ON hello
    FOR EACH ROW
    EXECUTE PROCEDURE myCoolFunction()
```

```python
%sql insert into hello (language, hello) VALUES ('brainfuck', '+++++++++++[>+++++[>++--------......----...++]');
```

```python
%sql DROP FUNCTION myCoolFunction CASCADE;
```
